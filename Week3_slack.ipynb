### Task: design slack-like system

#### 1. The top-level description of the proposed architecture;

Slack uses client-server architecture.

Key components of slack-like system:
- client (software on PCs and devices)
- webapp (manage and control)
- databases

#### 2. The detailed description of each architecture element.

- Client - any device that a person uses that requests access to a service provided by a server.

- Webapp - handle HTTP request/response cycles and communicate with other systems, data access management, contains the logic to look up metadata for a given chat/workspace, and then create a connection to the underlying database.

- Database - SQL database that stores the data on messages, chats, etc.

Slack also uses real-time message servers which sent the actual message after it was processed in webapp and stored.
